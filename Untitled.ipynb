```python
# plot season

#------------ helper functions
def convert_pr_units(clim):
    """ Function to convert precipitation units to mm/day
    
    Args:
        clim (dataset) : precipiation variable
        
    Output:
        clim (dataset) : converted variable and attributes
    """
    clim.data = clim.data * 86400
    clim.attrs['units'] = 'mm/day'
    return clim


def plot_climatology(clim,season,gridlines=False):
    """ Function to plot precipiation seasonal field
    
    Args:
        clim (dataset) : precipiation dataset
        season (string) : 3 Letter string for season e.g. 'JJA'
    
    """
    
    
    fig = plt.figure(figsize=[12,5])
    ax = fig.add_subplot(111, projection=ccrs.PlateCarree(central_longitude=180))
    clim.sel(season=season).plot.contourf(ax=ax,
                       levels=np.arange(0, 13.5, 1.5),
                       extend='max',
                       transform=ccrs.PlateCarree(),
                       cbar_kwargs={'label': clim.units},
                       cmap='cmo.haline_r')
    ax.coastlines()
    if gridlines:
        plt.gca().gridlines()
    ax.set_title(dset.attrs['model_id'] + ' precipitation climatology (' + season + ')')
    plt.show()
    
    

#-------------- MASTER function
def plot_pr_climatology(pr_file,season,gridlines=False):
    """ Plot the precipitation climatology
    
    Args:
        pr_file (str):Precipiation
        season (str): Season (3 letter abbreviation)
    """
    
    dset = xr.open_dataset(pr_file)
    clim = dset['pr'].groupby('time.season').mean('time', keep_attrs=True) 
    clim = convert_pr_units(clim)
    plot_climatology(clim,season,gridlines)

#
#
#--------------- call function
plot_pr_climatology(acces_pr_file,'DJF',gridlines=True)
```
